```python
df = pd.read_csv("./list_of_authors.csv")

df.head(10)
```

```python
authors = {}
authors['name'] = df['name'].values.tolist()
authors['email'] = df['email'].values.tolist()
authors['university'] = df['university'].values.tolist()
authors['scholar_id'] = []
```

```python
def institution(uni):
    mapping = {
        "SUST": "Shahjalal University of Science and Technology",
        "KUET": "Khulna University of Engineering and Technology",
        "RUET": "Rajshahi University of Engineering and Technology",
        "DUET": "Dhaka University of Engineering and Technology",
        "DU": "Dhaka University"
    } 
        
    return mapping[uni]
```

```python
count = 0

for naam in authors['name']:
    if type(naam) == str and naam != "":
        length = len(naam)
        
        searched = scholarly.search_author(naam)
        uni = authors['university'][count]
        uni = institution(uni)
        
        for s in searched:
            if uni in s['affiliation'] and len(s['name']) == length:
                authors['scholar_id'].append(s['scholar_id'])
                break
        
        if len(authors['scholar_id']) != count+1:
            authors['scholar_id'].append(None)
            
        if count%10 == 0:
            print(count, authors['scholar_id'][count])
    else:
        authors['scholar_id'].append(None)
    count = count + 1
   
```
